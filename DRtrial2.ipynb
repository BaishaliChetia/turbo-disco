```python
import numpy as np
from PIL import Image
import glob
import pandas as pd
import cv2
import time
import matplotlib.pyplot as plt
from multiprocessing import Pool
import os


def read_image_and_resize(fname: str, width:int=512, height:int=512):
  #print(f"Reading image and resizing for {fname} ({width}, {height})")
  img = cv2.imread(fname)
  img_resized = cv2.resize(img, (width, height,))
  img_resized = np.asarray(img_resized)
  img_resized = img_resized.reshape(tuple([1] + list(img_resized.shape)))
  img_resized = img_resized / 255.0
  img_resized = img_resized / np.max(img_resized)
  #print(f"Done processing {fname}")
  return [fname, img_resized]
  
def read_data(directory_glob:str, csv_filename:str)->tuple:
  X = None
  Y = np.zeros(0)
  height = 512
  width = 512

  # Load CSV as pandas
  y_pandas = pd.read_csv(csv_filename)

  directories = glob.glob(directory_glob)
  to_plot = 5

  begin = time.time()
  pool = Pool(os.cpu_count() * 4)
  images_and_fnames = pool.map(read_image_and_resize, directories[:100])

  for fname, img_resized in images_and_fnames:
    keyname = fname.split('.')[0].split('/')[1]
    y = y_pandas[y_pandas['image'] == keyname]['level']
    Y = np.append(Y, y)
    if X is None:
      X = img_resized
    else:
      X = np.concatenate((X, img_resized), axis=0)
 
  print(f"Everything too {time.time() - begin} seconds")
  return X, Y

X, Y = read_data("train/*", "trainLabels.csv")
print(X.shape, Y.shape)
print(Y)
```
